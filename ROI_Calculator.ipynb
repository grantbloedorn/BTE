```python
def roi(cost, revenue):
    if not isinstance(cost, (int, float)) or not isinstance(revenue, (int, float)):
        return "Error: Inputs must be numeric."
    if cost <= 0:
        return "Error: Cost must be greater than 0."
    return (revenue - cost) / cost

c = float(input("Enter cost: "))
r = float(input("Enter revenue: "))

result = roi(c, r)
print(result)
```
