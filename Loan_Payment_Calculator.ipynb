```python
def loan_payment(principal, annual_rate, years):
    if not isinstance(principal, (int, float)) or not isinstance(annual_rate, (int, float)) or not isinstance(years, (int, float)):
        return "Error: all inputs must be numbers."
    if principal <= 0 or annual_rate <= 0 or years <= 0:
        return "Error: all inputs must be positive."

    r = annual_rate / 12
    n = years * 12
    payment = (r * principal) / (1 - ((1 + r) ** (-n)))
    return payment

if __name__ == "__main__":
    p = float(input("Enter loan amount: "))
    rate = float(input("Enter annual interest rate (as decimal): "))
    yrs = float(input("Enter number of years: "))

    result = loan_payment(p, rate, yrs)
    print(result)
```
